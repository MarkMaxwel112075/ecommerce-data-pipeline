```python
import pandas as pd
import os
import sqlalchemy
import pymysql
import matplotlib
from dotenv import load_dotenv
from sqlalchemy import create_engine, text
from urllib.parse import quote_plus
from pathlib import Path
print("All libraries imported!")
```

```python
load_dotenv(override=True)
password = quote_plus(os.getenv('DB_PASSWORD'))
user     = os.getenv('DB_USER')
host     = os.getenv('DB_HOST')
port     = os.getenv('DB_PORT')
name     = os.getenv('DB_NAME')
conn_str = f'mysql+pymysql://{user}:{password}@{host}:{port}/{name}'
engine   = create_engine(conn_str)
with engine.connect() as conn:
    result = conn.execute(text("SELECT 'Connected to MySQL!' AS message"))
    for row in result:
        print(row[0])
```

```python
RAW_DIR = Path.home() / 'Desktop' / 'ecommerce_pipeline' / 'data' / 'raw'

print("Loading CSV files...")
orders    = pd.read_csv(RAW_DIR / 'olist_orders_dataset.csv')
items     = pd.read_csv(RAW_DIR / 'olist_order_items_dataset.csv')
customers = pd.read_csv(RAW_DIR / 'olist_customers_dataset.csv')
products  = pd.read_csv(RAW_DIR / 'olist_products_dataset.csv')
sellers   = pd.read_csv(RAW_DIR / 'olist_sellers_dataset.csv')
payments  = pd.read_csv(RAW_DIR / 'olist_order_payments_dataset.csv')
reviews   = pd.read_csv(RAW_DIR / 'olist_order_reviews_dataset.csv')
category  = pd.read_csv(RAW_DIR / 'product_category_name_translation.csv')

print(f"Orders:    {len(orders):,} rows")
print(f"Items:     {len(items):,} rows")
print(f"Customers: {len(customers):,} rows")
print(f"Products:  {len(products):,} rows")
print(f"Sellers:   {len(sellers):,} rows")
print(f"Payments:  {len(payments):,} rows")
print(f"Reviews:   {len(reviews):,} rows")
print(f"Category:  {len(category):,} rows")
print("\nAll files loaded successfully!")
```

```python
print("Cleaning orders...")
orders_clean = orders.copy()

# Convert date columns from text to datetime
date_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]
for col in date_cols:
    orders_clean[col] = pd.to_datetime(orders_clean[col], errors='coerce')

# Remove orders with no purchase timestamp
before = len(orders_clean)
orders_clean = orders_clean.dropna(subset=['order_purchase_timestamp'])
print(f"  Dropped {before - len(orders_clean)} orders with no timestamp")

# Add useful calculated columns
orders_clean['is_delivered'] = (orders_clean['order_status'] == 'delivered').astype(int)

orders_clean['delivery_days'] = (
    orders_clean['order_delivered_customer_date'] -
    orders_clean['order_purchase_timestamp']
).dt.days

orders_clean['is_late'] = (
    orders_clean['order_delivered_customer_date'] >
    orders_clean['order_estimated_delivery_date']
).astype(int)

orders_clean['order_month'] = (
    orders_clean['order_purchase_timestamp'].dt.to_period('M').astype(str)
)
orders_clean['order_year'] = orders_clean['order_purchase_timestamp'].dt.year

print(f"  Orders cleaned: {len(orders_clean):,} rows")
print(f"  Delivered orders: {orders_clean['is_delivered'].sum():,}")
print(f"  Late deliveries: {orders_clean['is_late'].sum():,}")
print("Orders cleaning done!")
```

```python
print("Cleaning products...")
products_clean = products.copy()

# Merge English category names
products_clean = products_clean.merge(
    category,
    on='product_category_name',
    how='left'
)

# Fill missing category names
products_clean['product_category_name_english'] = (
    products_clean['product_category_name_english'].fillna('unknown')
)

# Fill missing numeric columns with 0
numeric_cols = [
    'product_weight_g',
    'product_length_cm',
    'product_height_cm',
    'product_width_cm',
    'product_name_lenght',
    'product_description_lenght',
    'product_photos_qty'
]
products_clean[numeric_cols] = products_clean[numeric_cols].fillna(0)

# Rename columns to match MySQL table
products_clean = products_clean.rename(columns={
    'product_name_lenght':        'name_length',
    'product_description_lenght': 'description_length',
    'product_photos_qty':         'photos_qty',
    'product_weight_g':           'weight_g',
    'product_length_cm':          'length_cm',
    'product_height_cm':          'height_cm',
    'product_width_cm':           'width_cm'
})

print(f"  Products cleaned: {len(products_clean):,} rows")
print("Products cleaning done!")
```

```python
print("Cleaning order items...")
items_clean = items.copy()

# Convert shipping date
items_clean['shipping_limit_date'] = pd.to_datetime(
    items_clean['shipping_limit_date'], errors='coerce'
)

# Remove rows with zero or negative price
before = len(items_clean)
items_clean = items_clean[items_clean['price'] > 0]
print(f"  Dropped {before - len(items_clean)} items with invalid price")

# Fill missing freight with median
median_freight = items_clean['freight_value'].median()
items_clean['freight_value'] = items_clean['freight_value'].fillna(median_freight)

# Rename shipping column
items_clean = items_clean.rename(columns={
    'shipping_limit_date': 'shipping_date'
})

print(f"  Items cleaned: {len(items_clean):,} rows")
print("Items cleaning done!")
```

```python
print("Cleaning customers and sellers...")

customers_clean = customers.rename(columns={
    'customer_zip_code_prefix': 'zip_code'
})

sellers_clean = sellers.rename(columns={
    'seller_zip_code_prefix': 'zip_code'
})

print(f"  Customers: {len(customers_clean):,} rows")
print(f"  Sellers:   {len(sellers_clean):,} rows")
print("Customers and sellers cleaning done!")
```

```python
# Step 1 — Drop all tables in correct order
print("Dropping existing tables...")
with engine.connect() as conn:
    conn.execute(text("SET FOREIGN_KEY_CHECKS = 0"))
    conn.execute(text("DROP TABLE IF EXISTS order_reviews"))
    conn.execute(text("DROP TABLE IF EXISTS order_payments"))
    conn.execute(text("DROP TABLE IF EXISTS order_items"))
    conn.execute(text("DROP TABLE IF EXISTS orders"))
    conn.execute(text("DROP TABLE IF EXISTS products"))
    conn.execute(text("DROP TABLE IF EXISTS sellers"))
    conn.execute(text("DROP TABLE IF EXISTS customers"))
    conn.execute(text("SET FOREIGN_KEY_CHECKS = 1"))
    conn.commit()
print("All tables dropped!")
```

```python
# Step 2 — Recreate all tables with proper structure
print("Recreating tables...")
with engine.connect() as conn:
    conn.execute(text("""
        CREATE TABLE customers (
            customer_id         VARCHAR(50) PRIMARY KEY,
            customer_unique_id  VARCHAR(50),
            zip_code            VARCHAR(10),
            city                VARCHAR(100),
            state               CHAR(2)
        )
    """))
    conn.execute(text("""
        CREATE TABLE sellers (
            seller_id   VARCHAR(50) PRIMARY KEY,
            zip_code    VARCHAR(10),
            city        VARCHAR(100),
            state       CHAR(2)
        )
    """))
    conn.execute(text("""
        CREATE TABLE products (
            product_id              VARCHAR(50) PRIMARY KEY,
            category_name           VARCHAR(100),
            product_category_name_english VARCHAR(100),
            name_length             INT,
            description_length      INT,
            photos_qty              INT,
            weight_g                DECIMAL(10,2),
            length_cm               DECIMAL(10,2),
            height_cm               DECIMAL(10,2),
            width_cm                DECIMAL(10,2)
        )
    """))
    conn.execute(text("""
        CREATE TABLE orders (
            order_id                        VARCHAR(50) PRIMARY KEY,
            customer_id                     VARCHAR(50),
            order_status                    VARCHAR(20),
            order_purchase_timestamp        DATETIME,
            order_approved_at               DATETIME,
            order_delivered_carrier_date    DATETIME,
            order_delivered_customer_date   DATETIME,
            order_estimated_delivery_date   DATETIME,
            is_delivered                    TINYINT(1),
            delivery_days                   INT,
            is_late                         TINYINT(1),
            order_month                     VARCHAR(10),
            order_year                      INT
        )
    """))
    conn.execute(text("""
        CREATE TABLE order_items (
            order_id        VARCHAR(50),
            order_item_id   INT,
            product_id      VARCHAR(50),
            seller_id       VARCHAR(50),
            shipping_date   DATETIME,
            price           DECIMAL(10,2),
            freight_value   DECIMAL(10,2),
            PRIMARY KEY (order_id, order_item_id)
        )
    """))
    conn.execute(text("""
        CREATE TABLE order_payments (
            order_id                VARCHAR(50),
            payment_sequential      INT,
            payment_type            VARCHAR(30),
            payment_installments    INT,
            payment_value           DECIMAL(10,2),
            PRIMARY KEY (order_id, payment_sequential)
        )
    """))
    conn.execute(text("""
        CREATE TABLE order_reviews (
            review_id                   VARCHAR(50) PRIMARY KEY,
            order_id                    VARCHAR(50),
            review_score                INT,
            review_comment_title        VARCHAR(100),
            review_comment_message      TEXT,
            review_creation_date        DATETIME,
            review_answer_timestamp     DATETIME
        )
    """))
    conn.commit()
print("All tables recreated!")
```

```python
# Fix duplicate review_ids
print("Fixing duplicate reviews...")
reviews_cols = ['review_id', 'order_id', 'review_score',
                'review_comment_title', 'review_comment_message',
                'review_creation_date', 'review_answer_timestamp']

reviews_clean = reviews[reviews_cols].copy()

# Remove duplicates keeping first occurrence
before = len(reviews_clean)
reviews_clean = reviews_clean.drop_duplicates(subset=['review_id'], keep='first')
print(f"  Removed {before - len(reviews_clean)} duplicate reviews")
print(f"  Clean reviews: {len(reviews_clean):,} rows")

# Load to MySQL
reviews_clean.to_sql(
    name      = 'order_reviews',
    con       = engine,
    if_exists = 'append',
    index     = False,
    chunksize = 1000
)
print(f"  Loaded {len(reviews_clean):,} rows into order_reviews")
print("Reviews loaded successfully!")
```

```python
# Truncate reviews table and reload with duplicate handling
print("Clearing reviews table...")
with engine.connect() as conn:
    conn.execute(text("TRUNCATE TABLE order_reviews"))
    conn.commit()
print("Reviews table cleared!")

# Insert row by row ignoring duplicates
print("Loading reviews with duplicate handling...")
reviews_cols = ['review_id', 'order_id', 'review_score',
                'review_comment_title', 'review_comment_message',
                'review_creation_date', 'review_answer_timestamp']

reviews_clean = reviews[reviews_cols].drop_duplicates(
    subset=['review_id'], keep='first'
).reset_index(drop=True)

# Insert using INSERT IGNORE to skip any remaining duplicates
inserted = 0
skipped = 0
batch_size = 1000

with engine.connect() as conn:
    for i in range(0, len(reviews_clean), batch_size):
        batch = reviews_clean.iloc[i:i+batch_size]
        for _, row in batch.iterrows():
            try:
                conn.execute(text("""
                    INSERT IGNORE INTO order_reviews
                    (review_id, order_id, review_score,
                     review_comment_title, review_comment_message,
                     review_creation_date, review_answer_timestamp)
                    VALUES (:review_id, :order_id, :review_score,
                            :review_comment_title, :review_comment_message,
                            :review_creation_date, :review_answer_timestamp)
                """), row.to_dict())
                inserted += 1
            except:
                skipped += 1
        conn.commit()
        if (i // batch_size) % 10 == 0:
            print(f"  Progress: {i:,} / {len(reviews_clean):,} rows processed")

print(f"Reviews loaded: {inserted:,} inserted, {skipped:,} skipped")
print("Done!")
```

```python
# Clear and reload reviews properly
print("Clearing reviews table...")
with engine.connect() as conn:
    conn.execute(text("TRUNCATE TABLE order_reviews"))
    conn.commit()

# Get truly unique reviews
reviews_cols = ['review_id', 'order_id', 'review_score',
                'review_comment_title', 'review_comment_message',
                'review_creation_date', 'review_answer_timestamp']

reviews_final = reviews[reviews_cols].copy()

# Drop ALL duplicates including original — keep only truly unique review_ids
reviews_final = reviews_final.drop_duplicates(subset=['review_id'], keep='first')

# Also drop rows where review_id is null
reviews_final = reviews_final.dropna(subset=['review_id'])

print(f"Reviews to load: {len(reviews_final):,}")

# Load using replace mode — no primary key conflict possible
# since we dropped the table constraint temporarily
with engine.connect() as conn:
    conn.execute(text("ALTER TABLE order_reviews DROP PRIMARY KEY"))
    conn.commit()

reviews_final.to_sql(
    name      = 'order_reviews',
    con       = engine,
    if_exists = 'append',
    index     = False,
    chunksize = 5000
)

# Restore primary key
with engine.connect() as conn:
    conn.execute(text("ALTER TABLE order_reviews ADD PRIMARY KEY (review_id)"))
    conn.commit()

print(f"Loaded {len(reviews_final):,} rows into order_reviews")
print("Done!")
```

```python
print("Running analysis queries...")

monthly_revenue_sql = """
    SELECT
        o.order_month,
        COUNT(DISTINCT o.order_id)      AS total_orders,
        ROUND(SUM(oi.price), 2)         AS total_revenue,
        ROUND(SUM(oi.freight_value), 2) AS total_freight,
        ROUND(AVG(oi.price), 2)         AS avg_order_value
    FROM orders o
    JOIN order_items oi ON o.order_id = oi.order_id
    WHERE o.order_status = 'delivered'
    GROUP BY o.order_month
    ORDER BY o.order_month
"""

monthly_revenue = pd.read_sql(text(monthly_revenue_sql), engine)
print(f"Monthly revenue: {len(monthly_revenue)} months of data")
print(monthly_revenue.head(10).to_string(index=False))
```

```python
# Query 2 — Revenue by category
category_sql = """
    SELECT
        p.product_category_name_english  AS category,
        COUNT(DISTINCT o.order_id)        AS total_orders,
        ROUND(SUM(oi.price), 2)           AS total_revenue,
        ROUND(AVG(r.review_score), 2)     AS avg_review_score,
        ROUND(AVG(oi.freight_value), 2)   AS avg_freight
    FROM order_items oi
    JOIN orders o     ON oi.order_id   = o.order_id
    JOIN products p   ON oi.product_id = p.product_id
    LEFT JOIN order_reviews r ON o.order_id = r.order_id
    WHERE o.order_status = 'delivered'
    GROUP BY p.product_category_name_english
    ORDER BY total_revenue DESC
    LIMIT 15
"""
category_revenue = pd.read_sql(text(category_sql), engine)
print("TOP 15 CATEGORIES BY REVENUE")
print(category_revenue.to_string(index=False))
```

```python
# Query 3 — Revenue by state
state_sql = """
    SELECT
        c.state,
        COUNT(DISTINCT o.order_id)     AS total_orders,
        ROUND(SUM(oi.price), 2)        AS total_revenue,
        ROUND(AVG(oi.freight_value),2) AS avg_freight
    FROM orders o
    JOIN customers c    ON o.customer_id  = c.customer_id
    JOIN order_items oi ON o.order_id     = oi.order_id
    WHERE o.order_status = 'delivered'
    GROUP BY c.state
    ORDER BY total_revenue DESC
"""
state_revenue = pd.read_sql(text(state_sql), engine)
print("REVENUE BY STATE (top 10)")
print(state_revenue.head(10).to_string(index=False))

# Query 4 — Delivery performance
delivery_sql = """
    SELECT
        c.state,
        COUNT(o.order_id)                               AS total_orders,
        ROUND(AVG(o.delivery_days), 1)                  AS avg_delivery_days,
        ROUND(SUM(CASE WHEN o.is_late = 1
              THEN 1 ELSE 0 END) * 100.0
              / COUNT(o.order_id), 1)                   AS late_pct
    FROM orders o
    JOIN customers c ON o.customer_id = c.customer_id
    WHERE o.order_status = 'delivered'
    AND   o.delivery_days IS NOT NULL
    GROUP BY c.state
    ORDER BY late_pct DESC
    LIMIT 10
"""
delivery = pd.read_sql(text(delivery_sql), engine)
print("\nDELIVERY PERFORMANCE — worst states")
print(delivery.to_string(index=False))

# Query 5 — Payment methods
payment_sql = """
    SELECT
        payment_type,
        COUNT(*)                           AS total_transactions,
        ROUND(SUM(payment_value), 2)       AS total_value,
        ROUND(AVG(payment_installments),1) AS avg_installments
    FROM order_payments
    GROUP BY payment_type
    ORDER BY total_value DESC
"""
payments_df = pd.read_sql(text(payment_sql), engine)
print("\nPAYMENT METHODS")
print(payments_df.to_string(index=False))

# Query 6 — Review scores
review_sql = """
    SELECT
        review_score,
        COUNT(*)                              AS total_reviews,
        ROUND(COUNT(*) * 100.0
              / SUM(COUNT(*)) OVER(), 1)      AS pct_of_total
    FROM order_reviews
    GROUP BY review_score
    ORDER BY review_score DESC
"""
review_scores = pd.read_sql(text(review_sql), engine)
print("\nREVIEW SCORE DISTRIBUTION")
print(review_scores.to_string(index=False))

# Query 7 — Freight vs revenue
freight_sql = """
    SELECT
        p.product_category_name_english       AS category,
        ROUND(SUM(oi.price), 2)               AS total_revenue,
        ROUND(SUM(oi.freight_value), 2)       AS total_freight,
        ROUND(SUM(oi.freight_value) * 100.0
              / SUM(oi.price), 1)             AS freight_pct
    FROM order_items oi
    JOIN products p ON oi.product_id = p.product_id
    JOIN orders   o ON oi.order_id   = o.order_id
    WHERE o.order_status = 'delivered'
    GROUP BY p.product_category_name_english
    HAVING total_revenue > 10000
    ORDER BY freight_pct DESC
    LIMIT 12
"""
freight = pd.read_sql(text(freight_sql), engine)
print("\nFREIGHT COST AS % OF REVENUE — worst categories")
print(freight.to_string(index=False))

print("\nAll queries complete!")
```
